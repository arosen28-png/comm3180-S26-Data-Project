## Introduction

The University of Pennsylvania has remained one of the most influential institutions in the city of Philadelphia, but its impact can be seen far beyond campus boundaries. This project explores how Penn shapes University City differently from the rest of Philadelphia relfected by housing, business activity, and crime patterns. In short, this blog takes you through whether University City is truly a university-centered environment using public datasets on rent distribution, business licenses, and crime statistics.

Instead of viewing Penn as just a school located within the city we examine Penn as a force that has actively influenced the surrounding economic and social atmosphere.

---
### Research Questions

1. How do rent prices in University City compare to the rest of Philadelphia?
2. What types of businesses are most concentrated around the University of Pennsylvania?
3. How do crime patterns in University City differ from other Philadelphia neighborhoods?
4. To what extent does the University of Pennsylvania shape the economic and social environment of University City?

# Crime and Public Safety
![Penn Patrol Zone Violent Crime](penn_patrol_zone_crime.png)

One major area of analysis focused on crime patterns within the Penn Patrol Zone compared to broader citywide trends. While Philadelphia experienced significant increases in violent crime during the COVID-19 era, crime trends near Penn appeared comparatively stable.

The visualization above depicts violent crime incidents within the Penn Patrol Zone from 2017–2024.

The graph shows that robbery consistently remained one of the highest reported violent crimes near Penn, peaking in 2022 before declining by 2024. Aggravated assault fluctuated throughout the period, while rape and domestic violence remained comparatively lower.

Overall, the data suggests that while crime certainly exists within University City, the area did not experience the same dramatic upward trajectory seen across Philadelphia as a whole.

To provide more context, we also analyzed citywide homicide trends.

![Philadelphia Citywide Homicides](philadelphia_homicides.png)


Philadelphia homicides had a huge increase between 2020 and 2021, reaching a peak of 562 homicides in 2021 before gradually declining. Compared to to the Penn Patrol Zone visualization, this chart suggests that University City may operate differently from many other parts of Philadelphia due to more investment, policing, or university infrastructure.

# Business Composition Around Penn

Another major element of this project was understanding what kinds of businesses are found close to Penn.

![Business License Composition](business_license_composition.png)

The visualization shows a clear difference between business composition near Penn and outside University City. Areas within 1000 meters of Penn exhibit a wider range of business types such as food and dining establishments, operational services, and transportation-related businesses. 

Additionally, Businesses surrounding the university appear tailored toward student life and university-related demand whereas housing-related licenses are more dominate, the chart outside University City.

# Housing and Rent Distribution

Finally, looking into housing trends was pivitol for this analysis.

![Rent Distribution](rent_distribution.png)

The rent distribution graph reveals significant differences between University City and Philadelphia overall. University City contains a much larger concentration of renter households within the $1000–1249 range, while Philadelphia shows stronger concentrations in higher rent categories such as $1500–1999 and $2000+.

At the same time, University City also conists of more lower-cost rental categories which is likely because of student housing and shared apartments.


![Examples of rent prices in Upenn's campus](ucityprices.png)

Listings from https://offcampushousing.upenn.edu/listing

These findings tell us that University City's rental market is shaped by:

* Student demand
* Proximity to campus
* Dense apartment development

Rather than mirroring Philadelphia’s broader housing market, University City appears to have its own unique makeup.

# Final Takeaways

Across crime trends, business composition, and housing patterns, our findings suggest that University City functions differently from much of Philadelphia due to the concentrated influence of the University of Pennsylvania.

Penn’s presence appears to:

* Shape local business ecosystems
* Influence rental markets
* Affect neighborhood development patterns
* Contribute to distinct public safety dynamics
